```python
# Libraries
import pandas as pd
import numpy as np
import re
import sys

# Load File: 1998-2005 Data
# I converted the excel file to .csv format
file98_05 = pd.read_csv("QueryTrialInformation1998-2005.csv") 

# View Data
file98_05

type(file98_05)

# Create New Column: "Start Year" by extracting 4-digit year string from "Start Date"
search = []    
for values in file98_05['Start Date']:
    search.append(re.search(r'\d{4}', values).group())

file98_05['Start Year'] = search
file98_05

# Check unique entries in "Phases" Column
print(file98_05.Phases.unique())

# Create a new column "Phase_Updated" which only considers latest Phases, also this column only keeps numeric value
# "NaN" and "Not Applicable" are converted to"0"

file98_05['Phases_updated'] = file98_05['Phases'] 
file98_05.loc[file98_05["Phases_updated"] == "Phase 1", "Phases_updated"] = 1
file98_05.loc[file98_05["Phases_updated"] == "Phase 2", "Phases_updated"] = 2
file98_05.loc[file98_05["Phases_updated"] == "Phase 3", "Phases_updated"] = 3
file98_05.loc[file98_05["Phases_updated"] == "Phase 4", "Phases_updated"] = 4
file98_05.loc[file98_05["Phases_updated"] == "Early Phase 1", "Phases_updated"] = 1
file98_05.loc[file98_05["Phases_updated"] == "Phase 1|Phase 2", "Phases_updated"] = 2
file98_05.loc[file98_05["Phases_updated"] == "Phase 2|Phase 3", "Phases_updated"] = 3
file98_05.loc[file98_05["Phases_updated"] == "Not Applicable", "Phases_updated"] = 0
file98_05.loc[file98_05["Phases_updated"] == "nan", "Phases_updated"] = 0

# Drop rows if "Phases_updated" = 0 
# "0" corresponds to Nan or Not Applicable

file98_05 = file98_05[file98_05["Phases_updated"]>0]

# View Data
file98_05

# renaming "Sponsor/Collaboratos" to "Sponsor_Collaborators"
file98_05.rename(columns = {'Sponsor/Collaborators':'Sponsor_Collaborators'}, inplace = True)


# Check unique entries in "Sponsors/Collaborators" Column
unique_collabs = file98_05.Sponsor_Collaborators.unique()
type(unique_collabs)

print(unique_collabs[:])

# Print all Unique Collaborators
with np.printoptions(threshold=np.inf):
    print(unique_collabs)

np.count_nonzero(unique_collabs) # 1234 unique entries in Sponsor/Collaborators

# Split mutiple collaborators in separate columns
file98_05['colab1'] = file98_05["Sponsor_Collaborators"].str.split("|").str[0] # Collaborator 1
file98_05['colab2'] = file98_05["Sponsor_Collaborators"].str.split("|").str[1] # Collaborator 2
file98_05['colab3'] = file98_05["Sponsor_Collaborators"].str.split("|").str[2] # Collaborator 3
file98_05['colab4'] = file98_05["Sponsor_Collaborators"].str.split("|").str[3] # Collaborator 4

# View Data
file98_05

# Find unique collaboratos in the "colab1" column
unique_colab1=file98_05.colab1.unique()
unique_colab1.sort() # sort alphabetically 
with np.printoptions(threshold=np.inf):
    print(unique_colab1)

```
